Алгоритм минимального покрывающего дерева сначала строит на графе минимальное покрывающее дерево, а затем последовательно удаляет ребра с наибольшим весом.

```python
import numpy as np

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

from mst_clustering import MSTClustering
```

```python
# Генерируем данные
X, _ = make_blobs(n_samples=100, centers=3, cluster_std=1.0, random_state=42)
```

```python
# по порогу длины ребра
clusterer1 = MSTClustering(edge_threshold=2.0)
labels1 = clusterer1.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title(f'MST Clustering (threshold=2.0)\nКластеров: {len(np.unique(labels1))}')
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels1, cmap='viridis')
plt.title(f'MST Clustering (threshold=2.0)\nКластеров: {len(np.unique(labels1))}')
```

```python
# по числу кластеров
clusterer2 = MSTClustering(n_clusters=3)
labels2 = clusterer2.fit_predict(X)
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1])
plt.title(f'MST Clustering (n_clusters=3)')
```

```python
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=labels2, cmap='viridis')
plt.title(f'MST Clustering (n_clusters=3)')
```
